```python
#Linear Regression, Task 1
#Christian Johnston, CPSC 4383, Fall 2023

#The purpose of this code is to implement a linear regression to a set of data regarding the predction of a location
#for a food truck business using city and commercial data on food truck sales.

#For Task 1, I will use a tutorial from https://realpython.com/linear-regression-in-python/ to plot and analyze the data.

#for step 1, I will be importing the tools I will be using for this task.
#This includes numpy and scikit-learn as well as pandas.

import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt

#in Step 2 I will be importing my data using a basic data import of csv files implemented by pandas

data = pd.read_csv(r'C:\Users\Christian\Desktop\School Stuff\Fall 2023\AI\course content\ex1data1.csv')

#here i split the data into x and y lists for the regression model.

x = pd.DataFrame(data, columns=['Population'])
y = pd.DataFrame(data, columns=['Profit'])

#here i fit the data to the model using the linear regression method in scikit-learn which uses Ordinary Least Squares.
model = LinearRegression().fit(x,y)

#Here are some reference values provided for the coefficients and intercept.

print(f"coefficient of determination: {model.score(x,y)}")

print(f"intercept or b: {model.intercept_}")

print(f"slope or m: {model.coef_}")
#I am providing this for the plot of the linear regression line, it uses the predict method in LinearRegression.
#This shows where the prediction would lie in the statistical model.
yPrediction = model.predict(x)

#Here is a plot of the values and the least squares prediction line.
plt.scatter(x, y, c='k')
plt.title("Food Truck Profits Compared to City Population")
plt.xlabel("Population")
plt.ylabel("Profit")
plt.plot(x,yPrediction, 'r')
plt.show()
```
